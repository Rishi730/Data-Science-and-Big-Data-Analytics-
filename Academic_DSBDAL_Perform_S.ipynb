```python
np.random.seed(42)
data=np.random.exponential(scale=10,size=1000)
df_right=pd.DataFrame({'Original':data})
df_right['Log-Transform']=np.log1p(df_right['Original'])
df_right['Square-Root-Transform']=np.sqrt(df_right['Original'])
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
df_right['Min-Max-Transform']=scaler.fit_transform(df_right[['Original']])
```

```python

fig,axes=plt.subplots(2,2,figsize=(12,8))
sns.histplot(df_right['Original'],bins=30,kde=True,ax=axes[0,0])
sns.histplot(df_right['Log-Transform'],bins=30,kde=True,ax=axes[0,1])
sns.histplot(df_right['Square-Root-Transform'],bins=30,kde=True,ax=axes[1,0])
sns.histplot(df_right['Min-Max-Transform'],bins=30,kde=True,ax=axes[1,1])
```
